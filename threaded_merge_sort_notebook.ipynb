```python
from scripts.threaded_merge_sort import threaded_merge
import scripts.random_list as random_list
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import time
```

```python
arr = random_list.gen_list(10000)
threads = [1,2,4,8]
results = []
for t in threads:
	start_time = time.time()
	threaded_merge(arr, t)
	results.append(time.time() - start_time)

results
```

```python
plt.figure(figsize=(10,5))
plt.xticks(threads)
plt.title("Time it takes to sort a list with multi-threading using merge sort")
plt.xlabel("Threads")
plt.ylabel("Process duration (s)")
plt.tight_layout()
plt.grid()
plt.plot(threads, results)
plt.show()
```
